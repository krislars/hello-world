```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
from IPython.display import display, HTML
from IPython.html import widgets
from IPython.html.widgets import interact, interactive
```

```python
def plot_pendulum(theta0):
    dt = 0.01;
    tnum = 1000;
    t = np.arange(0,tnum*0.01,dt);
    theta = np.zeros(tnum);
    omega = np.zeros(tnum);
    
    theta[0]=theta0;
    omega[0]=0.0;
    for i in range(1,tnum):
        omega[i]=omega[i-1]-10.0*np.sin(theta[i-1])*dt;
        theta[i]=theta[i-1]+omega[i]*dt;
        
    fig = plt.figure(figsize = (8, 6));
    ax = fig.add_subplot(1,1,1);
    ax.plot(t, theta, 'r', linewidth=2,label='EulerCromer');
    ax.plot(t, theta0*np.cos((10.0**0.5)*t), 'b', linewidth=2, label='small angle')
    ax.set_xlabel('time')
    ax.set_ylabel('theta')
    ax.legend()
```

```python
interact(plot_pendulum, theta0=1.0);
```
